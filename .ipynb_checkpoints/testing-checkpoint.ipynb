```python
vocab_size = 10
projection_layer = 5
embedding_1 = nn.Embedding(vocab_size, projection_layer)
embedding_2 = nn.Embedding(vocab_size - 1, projection_layer)
```

```python
def forward_step(x_input, idx_path):
    '''
    skip-gram 중 random sample 하나 학습
    depth < max_depth
    input : x_input (idx, direction_list), idx_path(depth)
    '''

    x_idx = torch.Tensor([x_input[0]]).to(torch.long)
    dir_path = torch.Tensor(x_input[1]).to(torch.long)

    #(1,D) : hidden layer
    proj = embedding_1(x_idx)

    #(path_length, D)
    hirearchy_vectors = embedding_2(dir_path)

    #(path_lenth, 1)
    output = torch.matmul(proj, hirearchy_vectors.T)
    output = torch.sigmoid(output)
    
    return output
```

```python
l = [0,[3,2]]
i = [0,1]
i = torch.Tensor(i)
```
